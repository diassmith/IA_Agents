```sql
-- Criar base de dados
CREATE DATABASE IF NOT EXISTS BRONZE;
CREATE DATABASE IF NOT EXISTS SILVER

-- tabela TB_BITCOIN, dados brutos do bitcoin
CREATE TABLE IF NOT EXISTS BRONZE.TB_BITCOIN(
  data          STRING,
  open          DOUBLE,
  high          DOUBLE,
  low           DOUBLE,
  close         DOUBLE,
  volume        DOUBLE
)

-- drop table BRONZE.TB_BITCOIN

CREATE TABLE IF NOT EXISTS SILVER.TB_BITCOIN_WEEKLY(
  semana_inicio  DATE,
  semana_fim     DATE,
  abertura       DOUBLE,
  fechamento     DOUBLE,
  maxima         DOUBLE,
  minima         DOUBLE,
  variacao_pct   DOUBLE,
  volume_medio   DOUBLE
)


SELECT * FROM BRONZE.TB_BITCOIN;

SELECT * FROM SILVER.TB_BITCOIN
```
